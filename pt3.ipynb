##missing data##

```python
messy = pd.DataFrame({
    "product" :["pen","book","bag","lamp","moug"],
                     "price": [20,np.nan, 850,1200, np.nan],
"rating": [4.1,4.5,np.nan,4.5,3.9]
})               
```

```python
messy
```

```python
messy.isna().sum()
```

```python
filled = messy.copy()

```

```python
filled["price"] = filled["price"].fillna(filled["price"].mean())
```

```python
filled["rating"] = filled["rating"].fillna(filled["rating"].mean())
```

```python
print(filled)
```
